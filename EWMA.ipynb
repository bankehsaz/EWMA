```python
# import necessary libraries
import numpy as np
import pandas as pd
import plotly.express as px
import yfinance as yf
```

```python
# getting data from yahoo finance
df = yf.download(tickers='GLD', start='2023-01-01')
df.tail()
```

```python
# plotting the close price of SPDR Gold Shares (GLD)
px.line(df, x=df.index, y='Close')
```

```python
# Exponentially Weighted Moving Average Calculation
# Is EMA and EWMA the same?
# Wikipedia: Moving average page:
# An Exponential Moving Average (EMA), also known as an Exponentially Weighted Moving Average (EWMA)
df['EWMA-20'] = df['Close'].ewm(span=20).mean()
df.tail()
```

```python
# plotting the close price of SPDR Gold Shares (GLD) with EWMA-20
px.line(df, x=df.index, y=['Close', 'EWMA-20'])
```

```python
# Exponentially Weighted Moving Average Calculation
df['EWMA-60'] = df['Close'].ewm(span=60).mean()
df.tail()
```

```python
# plotting the close price of SPDR Gold Shares (GLD) with EWMA-20 and EWMA-60
px.line(df, x=df.index, y=['Close', 'EWMA-20', 'EWMA-60'])
```
